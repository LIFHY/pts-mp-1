```python
import os
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt

train_dir = "/content/drive/MyDrive/img classification_avenger/train"
validation_dir = "/content/drive/MyDrive/img classification_avenger/val"

img_size = (200, 200)
batch_size = 32

train_datagen = ImageDataGenerator(rescale=1.0 / 255.0)
val_datagen = ImageDataGenerator(rescale=1.0 / 255.0)

train_dataset = train_datagen.flow_from_directory(
    train_dir, target_size=img_size, batch_size=batch_size, class_mode="binary"
)

validation_dataset = val_datagen.flow_from_directory(
    validation_dir, target_size=img_size, batch_size=3, class_mode="binary"
)

print(train_dataset.classes)

model = tf.keras.models.Sequential(
    [
        tf.keras.layers.Conv2D(16, (3, 3), activation="relu", input_shape=(200, 200, 3)),
        tf.keras.layers.MaxPool2D(2, 2),
        tf.keras.layers.Conv2D(32, (3, 3), activation="relu"),
        tf.keras.layers.MaxPool2D(2, 2),
        tf.keras.layers.Conv2D(64, (3, 3), activation="relu"),
        tf.keras.layers.MaxPool2D(2, 2),
        tf.keras.layers.Flatten(),
        tf.keras.layers.Dense(512, activation="relu"),
        tf.keras.layers.Dense(1, activation="sigmoid"),
    ]
)

model.compile(loss="binary_crossentropy", optimizer=tf.keras.optimizers.RMSprop(learning_rate=0.001), metrics=["accuracy"])

model.fit(train_dataset, steps_per_epoch=3, epochs=10, validation_data=validation_dataset)
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

class DataAugmentor:
    def __init__(self, rescale=1.0 / 255.0):
        self.rescale = rescale
        self.augmentor = self._create_augmentor()

    def _create_augmentor(self):
        return ImageDataGenerator(
            rescale=self.rescale,
            rotation_range=30,
            width_shift_range=0.2,
            height_shift_range=0.2,
            shear_range=0.2,
            zoom_range=0.2,
            horizontal_flip=True,
            fill_mode="nearest",
        )

    def get_generator(self):
        return self.augmentor

train_data = DataAugmentor().get_generator()
```
